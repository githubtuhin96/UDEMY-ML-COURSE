```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',markers=["o",'v'],scatter_kws={'s':200})
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='sex',row='smoker')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='sex',hue='day',aspect=0.9,size=7)
```
