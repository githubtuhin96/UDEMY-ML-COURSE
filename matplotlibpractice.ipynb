```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
%matplotlib inline
```

```python
x=np.linspace(0,5,6)
y=x**2
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Title')
```

```python
plt.subplot(2,1,1)
plt.plot(x,y,'y')
plt.subplot(2,1,2)
plt.plot(y,x,'r')
```

```python
#object oriented method
fig=plt.figure()
axes=fig.add_axes([.1,.1,.8,.8])
axes.plot(x,y)
axes.set_xlabel('study')
axes.set_ylabel('success')
axes.set_title('sucess graph')
```

```python
fig =plt.figure()
axis1=fig.add_axes([.1,.1,.8,.8])
axis2=fig.add_axes([.25,.42,.3,.2])
axis1.plot(x,y)
axis1.set_title('Bigger one')
axis2.plot(x,y)
axis2.set_title('smallerone')
```
