```python
import seaborn as sns
%matplotlib inline
tips= sns.load_dataset('tips')
flight=sns.load_dataset('flights')
flight.head()
```

```python
cr=tips.corr()
```

```python
sns.heatmap(data=cr,cmap='coolwarm',annot=True)
```

```python
fl=flight.pivot_table(index='month',columns='year',values='passengers')
fl
```

```python
sns.heatmap(fl,cmap='coolwarm',linecolor='white',linewidth=1)
```

```python
sns.clustermap(fl,cmap='coolwarm',standard_scale=1)
```
