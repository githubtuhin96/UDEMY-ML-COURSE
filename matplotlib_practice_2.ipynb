```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
a=np.arange(10)
y=a**3
x=a**2
fig,axes=plt.subplots(nrows=5,ncols=2)
plt.tight_layout()
print(axes)
for elements in axes:
  for i in elements:
   i.plot(x,y)
```

```python
fig,axes=plt.subplots(nrows=5,ncols=2)
plt.tight_layout()
axes[0][1].set_title('2nd')
axes[0][1].plot(x,y)
```

```python
fig=plt.figure(figsize=(9,5))
ax=fig.add_axes([2,2,2,2])
ax.plot(x,y)
```

```python
fig,axes=plt.subplots(nrows=2,ncols=1,figsize=(8,8),dpi=558)
for e in axes:
  e.plot(x,y,'r',label='x^2')
  e.plot(y,x,'y',label='x^3')
  #e.legend(loc=10)
  e.legend(loc=(.2,.3))
plt.tight_layout()
```
